```python
import matplotlib.pyplot as plt
import numpy as np

# Example data set (ages)
ages = np.random.randint(18, 80, size=100)  # Generate 100 random ages between 18 and 80

# Create histogram
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=10, color='lightgreen', edgecolor='black')

# Add title and labels
plt.title('Distribution of Ages in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
